```python
class svm_classifier():
    def __init__(self,lr=0.1,plambda=0.01,epochs=50):
        self.w=None
        self.b=None
        self.lr=lr
        self.plambda=plambda
        self.epochs=epochs
    def fit(self,x,y):
        self.w=np.zeros(x.shape[1])
        self.b=0
        for j in range(self.epochs):
            yl=np.where(y<=0,-1,1)
            for i,xi in enumerate(x):
                condition=yl[i]*(np.dot(xi,self.w)-self.b)>=1
                if condition:
                    dw=2*self.plambda*self.w
                    db=0
                else:
                    dw=2*self.plambda*self.w-np.dot(xi,yl[i])
                    db = yl[i]
                self.w=self.w-(self.lr*dw)
                self.b=self.b-(self.lr*db)
    def predict(self,x):
        pred=np.dot(x,self.w)-self.b
        pl=np.sign(pred)
        yhat=np.where(pl<=-1,0,1)
        return yhat
```
