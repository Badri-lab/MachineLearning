#Low/Zero Variance or Constant features elimination:

```python
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
df=pd.read_csv('./CS570/data/santander.csv',nrows=10000)
print(df.columns)
x=df.drop(labels=['TARGET'],axis=1)
y=df.TARGET
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=2)
from sklearn.feature_selection import VarianceThreshold
vt=VarianceThreshold(0)
vt.fit(x_train)
constant_features=[c for c in x_train.columns if c not in x_train.columns[vt.get_support()]]
print('len(constant_features):',len(constant_features))
x_train=x_train.drop(constant_features,axis=1)
x_test=x_test.drop(constant_features,axis=1)
x_train.shape, x_test.shape
```

#Pearson Correlation to remove highly correlated INDEPENDENT features to prevent OVERFITTING:

```python
def correlation(dataset,threshold):
    col=set()
    df_corr=dataset.corr()
    for i in range(len(df_corr.columns)):
        for j in range(i):
            if abs(df_corr.iloc[i,j])>threshold:
                colname=df_corr.columns[i]
                col.add(colname)
    return col

import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
boston=pd.read_csv('.\CS570\data\BostonHousing.csv')
x=boston.drop(labels='MEDV',axis=1)
y=boston['MEDV']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=2)
print(x_train.shape,x_test.shape)
import seaborn as sns
plt.figure(figsize=(12,10))
df=x_train.corr()
#sns.heatmap(df,annot=True,cmap=plt.cm.CMRmap_r)
#plt.show()
corr_features=correlation(x_train,0.7)
print(corr_features)
x_train=x_train.drop(corr_features,axis=1)
x_test=x_test.drop(corr_features,axis=1)
print(x_train.shape,x_test.shape)
```
