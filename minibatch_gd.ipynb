```python
from sklearn.datasets import load_diabetes
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.model_selection import train_test_split
```

Getting Reference data using Ordinary Least Squares(OLS) method:

```python
x,y = load_diabetes(return_X_y=True)
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
lr=LinearRegression()
lr.fit(x_train,y_train)
print(lr.coef_)
print(lr.intercept_)
y_pred = lr.predict(x_test)
r2_score(y_test,y_pred)
#print("r2_score:",r2_score)
```

```python
class mbgdregressor:
    def __init__(self,batch_size,learn_rate=0.01,epochs=100):
        self.lr=learn_rate
        self.epochs=epochs
        self.coef_=None
        self.intercept_=None
        self.batch_size=batch_size

    def fit(self,x_train,y_train):
        #init the coefs and intercept
        self.intercept_=0
        self.coef_=np.ones(x_train.shape[1])

        for i in range(self.epochs):
            for j in range(int(x_train.shape[0]/self.batch_size)):
                #pick a batch of rows randomly
                idx=np.random.choice(x_train.shape[0],self.batch_size)
                
                #find y_hat for that row based on initialized values of coefs and intercept
                y_hat=np.dot(x_train[idx],self.coef_)+self.intercept_

                #Update intercept
                di=-2 * np.mean(y_train[idx] - y_hat)
                self.intercept_=self.intercept_- (self.lr * di)

                #update coefs
                dc = -2 * np.dot((y_train[idx] - y_hat),x_train[idx])
                self.coef_ = self.coef_ - (self.lr * dc)

        print(self.intercept_, self.coef_)

    def predict(self,x_test):
        return np.dot(x_test,self.coef_) + self.intercept_
```

```python
mb=mbgdregressor(batch_size=int(x_train.shape[0]/10),epochs=50,learn_rate=0.01) #epochs=100, learn_rate=0.8836568
mb.fit(x_train,y_train)
y_pred = mb.predict(x_test)
r2_score(y_test,y_pred)
```

```python
from sklearn.datasets import load_diabetes
from sklearn.linear_model import SGDRegressor
from sklearn.metrics import r2_score
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import numpy as np
x,y=load_diabetes(return_X_y=True)
xs=StandardScaler().fit_transform(x)
sgd_mb=SGDRegressor(max_iter=1,learning_rate='constant',eta0=0.0185,random_state=2,warm_start=True)
xs_train,xs_test,y_train,y_test=train_test_split(xs,y,test_size=0.2,random_state=2)
batch_size=32
epochs=20
n_samples=xs_train.shape[0]
for epoch in range(epochs):
    idx=np.random.permutation(n_samples)
    x_shuffled=xs[idx]
    y_shuffled=y_train[idx]
    for i in range(0,n_samples,batch_size):
        x_batch=x_shuffled[i:i+batch_size]
        y_batch=y_shuffled[i:i+batch_size]
        sgd_mb.partial_fit(x_batch,y_batch)
y_hat=sgd_mb.predict(xs_test)
print("Minibatch GD, r2_score",r2_score(y_test,y_hat))
print("Coeeficients: ",sgd_mb.coef_,sgd_mb.intercept_)
```
