```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

df = pd.read_csv("titanic.csv")
# print(df['Marital Status'].unique()) #list of unique values in column
```

# OneHotEncoding using sklearn

```python
ohe = OneHotEncoder(drop='first') #drop first to avoid perfect multicollinearity
odf = pd.DataFrame(
    ohe.fit_transform(df[["Marital Status"]]).toarray()
)  # perform OHE by calling fit_transform() and storing the result as a
dfm = pd.concat([df, odf], axis=1)  # vertically/columnar combine of 2 dataframes
dfm = dfm.drop(
    ["Marital Status"], axis=1
)  # dropping the source column of OHE dataframe
dfm.columns = [
    [
        "PassengerId",
        "Survived",
        "Pclass",
        "Name",
        "Sex",
        "Age",
        "SibSp",
        "Parch",
        "Ticket",
        "Fare",
        "Cabin",
        "Embarked",
        #"Divorced",
        "Married-civ-spouse",
        "Married-spouse-absent",
        "Never-Married",
        "Separated",
        "Widowed",
    ]
]  # check if this manual mapping can be automated
dfm
```

# OneHotEncoding using pandas.get_dummies() method.

```python
df_pgd = pd.concat([df, pd.get_dummies(df["Marital Status"], drop_first=True, dtype=int)], axis=1).drop(["Marital Status"], axis=1) #includes redundancy elimination fixing perfect multicollinearity
df_pgd
```

# Label Encoding without using sklearn

```python
df=pd.read_csv('titanic.csv')
#map() → works only on a Series(one field/column/domain), not on a DataFrame(tabular structure)
df['Survived']=df['Survived'].map({1:'yes',0:'no'}) #map to LabelEncode one column at a time
df[['Sex','Survived']]=df[['Sex','Survived']].replace({'male':0,'female':1,'yes':1,'no':0}) #replace to LabelEncode multiple columns in one shot
df
```

# A categorical column is said to have Nominal relationship when it has values logically unrelated.Ex: values present in Marital Status column.
# Ordinal relationship when it has values that are logically related. Ex: months of year, days of week, cancer type: Stage1, Stage2, Stage3, Stage4. Blood Pressure: Low, Medium, High
# When there only 2 values in a categorical column we use Label Encoding.
# Label Encoding can be applied on categorical columns with ordinal relationship or a categorical column with only 2 values.
# We apply OneHotEncoding when a categorical column had values with nominal relationships.
Practical rule of thumb
# If categories have order → use OrdinalEncoder
# If categories are just labels → use OneHotEncoder
# If encoding target variable → use LabelEncoder

```python
# ColumnTransformer
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OrdinalEncoder, OneHotEncoder
#quality is the name of the column where ordinal encoding is applied
#Sex, Embarked are the columns on which OHE is applied
preprocessor = ColumnTransformer([
    ('ord', OrdinalEncoder(categories=[['low','medium','high']]), ['quality']),
    ('nom', OneHotEncoder(), ['Sex', 'Embarked'])
])
```

```python
import pandas as pd

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LogisticRegression

# Load data
df = pd.read_csv('titanic.csv')

# Target (keep it numeric — no need to convert to yes/no)
y = df['Survived']

# Features
X = df.drop(columns=['Survived'])

# Column groups
categorical_cols = ['Sex', 'Embarked']
numeric_cols = ['Age', 'Fare', 'SibSp', 'Parch', 'Pclass']

# --- Pipelines for each type ---

# Categorical pipeline
cat_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('encoder', OneHotEncoder(drop='first', handle_unknown='ignore'))
])

# Numeric pipeline
num_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

# --- ColumnTransformer ---
preprocessor = ColumnTransformer([
    ('cat', cat_pipeline, categorical_cols),
    ('num', num_pipeline, numeric_cols)
])

# --- Full pipeline ---
model = Pipeline([
    ('preprocessing', preprocessor),
    ('classifier', LogisticRegression())
])

# Fit model
model.fit(X, y)
```
